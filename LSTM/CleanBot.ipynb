```python
import urllib.request
urllib.request.urlretrieve('https://raw.githubusercontent.com/bab2min/corpus/master/sentiment/naver_shopping.txt', 'shopping.txt')
```

```python
import pandas as pd
import numpy as np

raw = pd.read_table('shopping.txt', names = ['rating', 'review'])
print(raw)

# 3보다 작으면 1로 표시 아니면 0으로
raw['label'] = np.where(raw['rating']>3,1,0)
print(raw)
```

```python
# Data 전처리
raw['review'] = raw['review'].str.replace('[^ㄱ-ㅎㅏ-ㅣ가-힣0-9 ]','')
# print(raw.isnull().sum())

raw.drop_duplicates(subset=['review'], inplace=True)
#print(raw)

# bag of words
unq_str = raw['review'].tolist()
unq_str = ''.join(unq_str)
unq_str = list(set(unq_str))
unq_str.sort()
print(unq_str[0:500])

```

```python
from tensorflow.keras.preprocessing.text import Tokenizer

# 문자들을 정수로 (char_level 글자단위를 정수로),(oov_token 처음보는 단어는 어떻게 바꿀지 정의)
tokenizer = Tokenizer(char_level=True, oov_token = '<OOV>')

# 딕셔너리 형태
str_list = raw['review'].tolist()
tokenizer.fit_on_texts(str_list)
print(tokenizer.word_index)

train_seq = tokenizer.texts_to_sequences(str_list)
print(train_seq[0:10])

Y = raw['label'].tolist()
print(Y[0:10])
```

```python
raw['length'] = raw['review'].str.len()

print(raw.head())
print(raw.describe())

raw['length'][raw['length']<120].count()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

# 120자 미만 글자들 0으로 채워주기
X = pad_sequences(train_seq, maxlen=120)

# 데이터 쪼개기
trainX, valX, trainY, valY = train_test_split(X,Y,test_size=0.2,random_state=42)
print(len(trainX))
print(trainX[:1])
print(trainY[:100])
print(len(valX))
print(np.array(trainX).shape)
print(np.array(trainY).shape)
```

```python
import tensorflow as tf

# trainY를 원-핫 인코딩
trainY_encoded = tf.one_hot(trainY, depth=2)

model = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(len(unq_str) + 1, 16),
    tf.keras.layers.LSTM(100, input_shape=(50, 16)),
    tf.keras.layers.Dense(2, activation='softmax')
])
model.summary()
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['acc'])
model.fit(trainX, trainY_encoded, batch_size=64, epochs=5)
model.save('CleanBotModel')
```

```python
Pmodel = tf.keras.models.load_model('CleanBotModel')

# 새로운 데이터 (입력 데이터) 준비
new_data = ['이거 완전 좋아']
test_seq = tokenizer.texts_to_sequences(new_data)
testX = pad_sequences(test_seq, maxlen=120)

# 새로운 데이터에 대한 예측
predictions = Pmodel.predict(testX)

print(f"부정 : {predictions[0][0]}, 긍정 : {predictions[0][1]}")


```
